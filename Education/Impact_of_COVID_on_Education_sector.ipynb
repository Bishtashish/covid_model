## Impact of COVID-19 on Education sector

**In this notebook, we have used Tables for COVID-19 Pandemic Impact on How Children Received Education in NC state from 'Household Pulse Survey Data Tables for 2020'**

**Attribute Details:**
- No of people for whom Classes were moved to a distance learning format - online
- No of people for whom Classes were moved to a distance learning format - paper sent to home
- No of people for whom classes were cancelled	
- No of people for whom classes changed in another way	
- No of people for whom no change to classes because schools did not close
- Total no of people who participated in the survey

**Key Findings:**
- Most of the education institutes/universities moved to Online distance Learning Format at the end of Spring Semester when NC state declared phase 1 of stay at home order on May 8 effective till May 22, 2020.
- A few education institutes/universities preferred paper material sent to students.
- Most of education instities/universities changed their format from in-person to Online at the start of Fall semester so percentage of Class format change observed spike at the beginning of Fall in week 13.
- And as Most universities preffered Online distance learning format, the class cancelletion percentage decreased at the beggining of Fall.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
```

```python
df_education_medium = pd.read_csv('impact_on_education_received_medium.csv',na_values='-')
```

```python
df_education_medium.info()
```

```python
df_education_medium['Where no change to classes becauseschools did not close']
```

```python
df_education_medium['Did not report']
```

```python
df_education_medium = df_education_medium.drop('Unnamed: 0',axis=1)
df_education_medium
```

```python
# NA values replaced with 0
df_education_medium = df_education_medium.fillna(0)
```

```python
df_education_medium['%online_resources'] = df_education_medium.iloc[:,1]/df_education_medium.iloc[:,0] *100
df_education_medium['%paper_mat_sent_home'] = df_education_medium.iloc[:,2]/df_education_medium.iloc[:,0] *100
df_education_medium['%classes_canceled'] = df_education_medium.iloc[:,3]/df_education_medium.iloc[:,0] *100
df_education_medium['%class_format_changed'] = df_education_medium.iloc[:,4]/df_education_medium.iloc[:,0] *100
df_education_medium['%no_change_in_class'] = df_education_medium.iloc[:,5]/df_education_medium.iloc[:,0] *100
df_education_medium['%did_not_report'] = df_education_medium.iloc[:,6]/df_education_medium.iloc[:,0] *100
df_education_medium
```

```python
df_education_medium.set_index('week')
```

```python
style.use('ggplot')
plt.plot(df_education_medium['week'],df_education_medium['%online_resources'], label='online resources')
plt.xticks(range(1,19))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('% usage of online resources')
plt.title('Effect of COVID-19 on usage of online resources')
plt.legend()
```

## Observations:
**When NC state declared phase 1 stay at home order on May 8, the usage of online resources sky rocketed.**
Towards end of school in Spring'20 in week 2 (May 7 to May 12), the usage of online resources for education purpose drastically increased from 73% to 87%

**During phase 2 of stay at home order in NC (between May 20 to sept 1), the usage was in range 68% to 78%** Usage of online resources are lower in phase 2  compared to phase 1, due to summer vacation


```python
style.use('ggplot')
plt.plot(df_education_medium['week'],df_education_medium['%paper_mat_sent_home'])
plt.xticks(range(1,19))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('% paper material sent home')
plt.title('Effect of COVID-19 on sending paper material to students')
```

## Observations:

**Percentage of sending paper material to students kept on increasing from week 1 to week 10 (April 23 to July 27). And in the beginning of fall in week 13, it is drastically decreased.**

```python
style.use('ggplot')
plt.plot(df_education_medium['week'],df_education_medium['%classes_canceled'])
plt.xticks(range(1,19))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('% classes canceled')
plt.title('% classes canceled due to COVID-19')
```

## Observations:
**In NC state phase 1 and phase 2 stay at home order during May 8 to sept 1, percentage of class cancellation increased upto 50 %. Afterwards it decreased and spiked again in week 16 (Sept 30 to Oct 12).  The spike in week 16 might be caused due to many universities were first decided to start in-person classes in October but again decided to resume online classes.**

```python
style.use('ggplot')
plt.plot(df_education_medium['week'],df_education_medium['%class_format_changed'])
plt.xticks(range(1,19))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('%class_format_changed')
plt.title('% classes fomat changed due to COVID-19')
```

## Observstions:
**There was a little increase in Percentage of class format change in week2 in phase 1 stay at home order but it spiked at the beginning of Fall in week 13 (Aug 19 to Aug 31) and kept on increasing upto week 17 (October 14 to October 26).**

```python
style.use('ggplot')
plt.plot(df_education_medium['week'],df_education_medium['%no_change_in_class'])
plt.xticks(range(1,19))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('%no_change_in_class')
plt.title('% No change in class format')
```

```python
style.use('ggplot')
plt.plot(df_education_medium['week'],df_education_medium['%paper_mat_sent_home'],label='paper material sent home')
plt.plot(df_education_medium['week'],df_education_medium['%classes_canceled'],label='classes canceled')
plt.plot(df_education_medium['week'],df_education_medium['%class_format_changed'],label='class format changed')
plt.xticks(range(1,19))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('%')
plt.title('Effect of COVID-19 on educational resources')
plt.legend()
```

## Observations:
**In week 13 (Aug 19 to Aug 31) i.e. beginning of Fall, the percentage of class format changed increased. As a result, the percentage of class cancellations decreased and also usage of paper material sent to home decreased.**

```python
plt.plot(df_education_medium['week'],df_education_medium['%classes_canceled'],label='classes cancelled')
plt.plot(df_education_medium['week'],df_education_medium['%class_format_changed'],label='class format changed')
plt.xticks(range(1,19))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('%')
plt.title('Effect of COVID-19 on educational resources')
plt.legend()
```

## Observations:
**At the beginning of Fall in week 13, the percentage of class cancellation decreased and percentage of class format change increased as most of the universities decided to go online from in-person classe format.**

```python
import os #to get paths of excel
import pandas as pd
import matplotlib.pyplot as plt

# function to get all file paths from provided directory
def get_list_of_files(dir_path):
    # create a list of file and sub directories 
    # names in the given directory 
    files = os.listdir(dir_path)
    all_files = list()
    # Iterate over all the entries
    for file in files:
        # Create full path
        fullPath = os.path.join(dir_path, file)
        # If entry is a directory then get the list of files in this directory 
        if os.path.isdir(fullPath):
            all_files = all_files + get_list_of_files(fullPath)
        else:
            all_files.append(fullPath)
                
    return all_files


def get_folder_name_list(dir_path):
    base_path_folder_list=[]
    for entry in os.listdir(dir_path):
        if os.path.isdir(os.path.join(dir_path, entry)):
            base_path_folder_list.append(entry) 
    return base_path_folder_list

base_path = os.path.dirname(os.path.abspath("__file__"))

folders = get_folder_name_list(base_path)
print('folder name: '+ str(folders))


## impact_on_post_secondary_education_plan_total_count
## weeks : 13 to 18

xl_file_paths = get_list_of_files(base_path + '\\health1')
print('file paths: '+ str(xl_file_paths))

health_df = pd.DataFrame()

no_of_weeks = len(xl_file_paths)
col_names = ['idx', 'Total',
       'Received counseling or therapy from a mental health professional such as a psychiatrist, psychologist, psychiatric nurse, or clinical social worker',
     'Did not Receive counseling or therapy from a mental health professional ' , 'Did not Report counseling or therapy from a mental health professional ' ,
       'Needed counseling or therapy from a mental health professional, but did not get it for any reason',
             'No, Needed counseling or therapy from a mental health professional, but did not get it for any reason' ,      
       'Did not report, Needed counseling or therapy from a mental health professional, but did not get it for any reason',
                 
       'Took prescription medication to help with any emotions or with your concentration, behavior or mental health',
            'No, Took prescription medication to help with any emotions or with your concentration, behavior or mental health',
            'Did not report, Took prescription medication to help with any emotions or with your concentration, behavior or mental health']
idx = 18 - no_of_weeks + 1
print(no_of_weeks)
for file_path in xl_file_paths :
    # open file
    #print(file_path)
    week = file_path.split('\\')
    '''
    if week[6].split('.')[0].split('_')[1]=='se':
        path = os.path.join(base_path+ '\\health1', week[6])  
    
# Remove the file  
# 'file.txt'  

        os.remove(path)
    else:
      '''  
    week_number = week[6].split('.')[0].split('_')[1].replace('week','')
    
    df_temp = pd.read_excel(file_path, sheet_name='NC',index_col=0,names=col_names,usecols="A:K", )
    
    df_temp['week'] = int(week_number) 
    row_for_total_count = df_temp.loc[['Total']] #total
    #print(row_for_total_count)
    #append first row 'total' to final dataframe
    health_df = health_df.append(row_for_total_count,ignore_index=True)
    
health_df = health_df.sort_values(by=['week'],ascending=True)
print(health_df)
# create a CSV file 
health_df.to_csv(base_path+'\\impact_on_health_sector.csv')
```


In this notebook, we have used Tables for COVID-19 Pandemic Impact on How health sector is affected NC state from 'Household Pulse Survey Data Tables for 2020'

Attribute Details:

No of people for who Received counseling or therapy from a mental health professional such as a psychiatrist, psychologist, psychiatric nurse, or clinical social worker

No of people for who Did not Receive counseling or therapy from a mental health professional such as a psychiatrist, psychologist, psychiatric nurse, or clinical social worker

No of people for who Did not Report counseling or therapy from a mental health professional such as a psychiatrist, psychologist, psychiatric nurse, or clinical social worker

No of people who Needed counseling or therapy from a mental health professional, but did not get it for any reason

No of people who did not Needed counseling or therapy from a mental health professional, but did not get it for any reason

No of people who did not report counseling or therapy from a mental health professional, but did not get it for any reason

No of people who Took prescription medication to help with any emotions or with your concentration, behavior or mental health

No of people who did not Took prescription medication to help with any emotions or with your concentration, behavior or mental health

No of people who did not report prescription medication to help with any emotions or with your concentration, behavior or mental health

```python
import os #to get paths of excel
import pandas as pd
import matplotlib.pyplot as plt

df_health=pd.read_csv('impact_on_health_sector.csv',na_values='-')
```

```python
df_health
```

```python
df_health = df_health.drop('Unnamed: 0',axis=1)
```

```python
df_health
```

```python
df_health['%Received counseling or therapy'] =df_health.iloc[:,1]/df_health.iloc[:,0] *100
df_health['%Did not Receive counseling or therapy'] = df_health.iloc[:,2]/df_health.iloc[:,0] *100
df_health['%Did not Report counseling or therapy'] = df_health.iloc[:,3]/df_health.iloc[:,0] *100
df_health['%Needed counseling or therapy'] = df_health.iloc[:,4]/df_health.iloc[:,0] *100
df_health['%Not Needed counseling or therapy'] = df_health.iloc[:,5]/df_health.iloc[:,0] *100
df_health['%Did not report, Needed counseling or therapy'] = df_health.iloc[:,6]/df_health.iloc[:,0] *100
df_health['%Took prescription medication to help with any emotions'] = df_health.iloc[:,7]/df_health.iloc[:,0] *100
df_health['%Did not Take prescription medication to help with any emotions'] = df_health.iloc[:,8]/df_health.iloc[:,0] *100
df_health['%Did not report, prescription medication to help with any emotions'] = df_health.iloc[:,9]/df_health.iloc[:,0] *100
df_health
```

```python
df_health.set_index('week')
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as style
```

```python
style.use('ggplot')
plt.plot(df_health['week'],df_health['%Received counseling or therapy'], label='Received counseling or therapy')
plt.xticks(range(13,19))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('% Received counseling or therapy')
plt.title('Effect of COVID-19 on health for Received counseling or therapy')
plt.legend()
```

Observations:
When NC state declared stay at home order, the people were confused and had no idea of the extent of pandemic imimpact. Towards week 15-17 most of the people seeked mental help, this was at peak on week 16 for about 8.3% of the total number.



```python
style.use('ggplot')
plt.plot(df_health['week'],df_health['%Did not Receive counseling or therapy'], label='Did not Receive counseling or therapy')
plt.xticks(range(13,19))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('%Did not Receive counseling or therapy')
plt.title('Effect of COVID-19 on health who did not Receive counseling or therapy')
plt.legend()
```

Observations:

From week 13-15 most of the people did not seek mental help, there is a downfall on week 16 for about 70% of the total number, this explains the previous plot of the people who seekd for help.

## Impact of COVID-19 on Education sector

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
```

```python
df_education_medium = pd.read_csv('impact_on_education_received_medium.csv',na_values='-')
```

```python
df_education_medium.info()
```

```python
df_education_medium['Where no change to classes becauseschools did not close']
```

```python
df_education_medium['Did not report']
```

```python
df_education_medium = df_education_medium.drop('Unnamed: 0',axis=1)
df_education_medium
```

```python
df_education_medium = df_education_medium.fillna(0)
```

```python
df_education_medium['%online_resources'] = df_education_medium.iloc[:,1]/df_education_medium.iloc[:,0] *100
df_education_medium['%paper_mat_sent_home'] = df_education_medium.iloc[:,2]/df_education_medium.iloc[:,0] *100
df_education_medium['%classes_canceled'] = df_education_medium.iloc[:,3]/df_education_medium.iloc[:,0] *100
df_education_medium['%class_format_changed'] = df_education_medium.iloc[:,4]/df_education_medium.iloc[:,0] *100
df_education_medium['%no_change_in_class'] = df_education_medium.iloc[:,5]/df_education_medium.iloc[:,0] *100
df_education_medium['%did_not_report'] = df_education_medium.iloc[:,6]/df_education_medium.iloc[:,0] *100
df_education_medium
```

```python
df_education_medium.set_index('week')
```

```python
style.use('ggplot')
plt.plot(df_education_medium['week'],df_education_medium['%online_resources'], label='online resources')
plt.xticks(range(1,19))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('% usage of online resources')
plt.title('Effect of COVID-19 on usage of online resources')
plt.legend()
```

```python
style.use('ggplot')
plt.plot(df_education_medium['week'],df_education_medium['%paper_mat_sent_home'])
plt.xticks(range(1,19))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('% paper material sent home')
plt.title('Effect of COVID-19 on sending paper material to students')
```

```python
style.use('ggplot')
plt.plot(df_education_medium['week'],df_education_medium['%classes_canceled'])
plt.xticks(range(1,19))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('% classes canceled')
plt.title('% classes canceled due to COVID-19')
```

```python
style.use('ggplot')
plt.plot(df_education_medium['week'],df_education_medium['%class_format_changed'])
plt.xticks(range(1,19))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('%class_format_changed')
plt.title('% classes fomat changed due to COVID-19')
```

```python
style.use('ggplot')
plt.plot(df_education_medium['week'],df_education_medium['%no_change_in_class'])
plt.xticks(range(1,19))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('%no_change_in_class')
plt.title('% No change in class format')
```

```python
style.use('seaborn-poster')
plt.plot(df_education_medium['week'],df_education_medium['%online_resources'], label='online resources')
plt.plot(df_education_medium['week'],df_education_medium['%paper_mat_sent_home'],label='paper material sent home')
plt.plot(df_education_medium['week'],df_education_medium['%classes_canceled'],label='classes canceled')
plt.plot(df_education_medium['week'],df_education_medium['%class_format_changed'],label='class format changed')
plt.plot(df_education_medium['week'],df_education_medium['%no_change_in_class'],label='no change in class')
plt.xticks(range(1,19))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('% usage of online resources')
plt.title('Effect of COVID-19 on usage of online resources')
plt.legend()
```

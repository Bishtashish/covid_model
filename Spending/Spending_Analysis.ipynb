```python
style.use('ggplot')
plt.plot(df_spending['week'],df_spending['%Avoided eating at restaurants'], label='Avoided eating at restaurants')
plt.xticks(range(13,18))
plt.xlabel('Week Number')
plt.grid(True)
plt.ylabel('% Avoided eating at restaurants')
plt.title('Effect of COVID-19 on Avoided eating at restaurants')
plt.legend()
```

Observations:

From the above plot we can observe that there was a drastic decrease from week 13 to week 15 when after the breakout of Covid-19 pandemic. It shows that more number of people avoided going to the restaurants at that time. We can observe that from week 15 to week 16 there was an increase in the number of people going to the restaurants. There was a decrease observed from week 16 to week 17 and 
then a sudden rise from week 17 can be observed. This means that after week 17 more number of people avoided going to the restaurants.

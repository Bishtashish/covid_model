```python
from bs4 import BeautifulSoup
import requests
```

```python
# downloading files for impact_on_post_secondary_education_plan
files_to_download = []        
for i in range(18):
    url = 'https://www.census.gov/data/tables/2020/demo/hhp/hhp'+str(i+1)+'.html'
    #print(url)
    html_text = requests.get(url).text
    soup = BeautifulSoup(html_text,'lxml')
    anchors = soup.find_all('a', {'class': 'uscb-layout-align-start-start', 'href': True})
    for anchor in anchors:
        if(anchor['href'].find('spending2_')>-1 and anchor['href'].find('spending2_se')==-1):
            files_to_download.append(anchor['href'])
print(files_to_download)
base_path = '/Users/aakankshachauhan/KDD_Project_Fall_2020'
for file_path in files_to_download:
    dls = 'http:'+file_path
    file_name = file_path.split('/')
    output_file_path = base_path + file_name[9]
    resp = requests.get(dls)
    output = open(output_file_path, 'wb')
    output.write(resp.content)
    output.close()
```

```python
# -*- coding: utf-8 -*-
"""
Created on Wed Nov 25 13:46:57 2020

@author: AC
"""


import os #to get paths of excel
import pandas as pd
import matplotlib.pyplot as plt

# function to get all file paths from provided directory
def get_list_of_files(dir_path):
    # create a list of file and sub directories 
    # names in the given directory 
    files = os.listdir(dir_path)
    all_files = list()
    # Iterate over all the entries
    for file in files:
        print(file)
        # Create full path
        if file.find('xlsx')>-1:
            print("-----")
            fullPath = os.path.join(dir_path, file)
            # If entry is a directory then get the list of files in this directory 
            if os.path.isdir(fullPath):
                all_files = all_files + get_list_of_files(fullPath)
            else:
                all_files.append(fullPath)
                
    return all_files


def get_folder_name_list(dir_path):
   base_path_folder_list=[]
   for entry in os.listdir(dir_path):
       if os.path.isdir(os.path.join(dir_path, entry)):
           base_path_folder_list.append(entry) 
   return base_path_folder_list

base_path = '/Users/aakankshachauhan/KDD_Project_Fall_2020/spending2/'

# folders = get_folder_name_list(base_path)
# print('folder name: '+ str(folders))


## impact_on_spending_and_shopping_behavior_total_count
## weeks : 13 to 18

xl_file_paths = get_list_of_files(base_path)
print('file paths: '+ str(xl_file_paths))

spend_df = pd.DataFrame()

no_of_weeks = len(xl_file_paths)
col_names = ['Select characteristics', 'Total*',
       'Made more purchases online',
       'Made more purchases curbside pickup',
       'Made more purchases in-store',
       'Increased use of credit cards or smartphone apps for purchases',
       'Increased use of cash instead of credit cards or smartphone apps for purchases',
       'Avoided eating at restaurants',
       'Resumed eating at restaurants',
       'Canceled or postponed in-person medical or dental appointments',
       'Attended in-person medical or dental appointments',
       'Canceled or postponed housekeeping or caregiving services',
       'Resumed or started new housekeeping or caregiving services',
       'Did not make any changes to spending or shopping behavior',
        'Did not report' ]
idx = 18 - no_of_weeks + 1
print(no_of_weeks)
for file_path in xl_file_paths :
    # open file
    print(file_path)
    
    path, file = os.path.split(file_path)
    print(file)
    print(path)
    
    week_number = file.split('.')[0].split('_')[1].replace('week','')
    print(week_number)
    df_temp = pd.read_excel(file_path, sheet_name='NC',index_col=0,names=col_names,usecols="A:O", )
    
    df_temp['week'] = int(week_number)
    row_for_total_count = df_temp.loc[['Total']] #total
    #print(row_for_total_count)
    #append first row 'total' to final dataframe
    spend_df = spend_df.append(row_for_total_count,ignore_index=True)
    
spend_df = spend_df.sort_values(by=['week'],ascending=True)
print(spend_df)
# create a CSV file 
spend_df.to_csv('/Users/aakankshachauhan/KDD_Project_Fall_2020/spending2_final.csv')
```
